# Message Content Caching — InvokeModel API (Anthropic)

This notebook demonstrates how to cache message content using the **InvokeModel** and **InvokeModelWithResponseStream** APIs with Anthropic Claude's native `cache_control` syntax.

## What you'll learn

- **InvokeModel API**: `cache_control` embedded in each content block
- **InvokeModelWithResponseStream**: Same syntax with streaming response parsing
- **Simplified mode**: Single cache checkpoint covering all static content
- **Multiple checkpoints**: Granular caching of individual content sections

## Anthropic-specific syntax

```python
{"type": "text", "text": "<content>", "cache_control": {"type": "ephemeral", "ttl": "5m"}}
```

For the model-agnostic Converse API syntax (`cachePoint`), see [converse_api/](../../converse_api/).

## Prerequisites

- AWS credentials configured (default profile)
- Access to Claude Sonnet 4.6 on Amazon Bedrock
- `boto3 >= 1.42.0`

## Token threshold

Claude Sonnet 4.6 requires at least **2,048 tokens** per cache checkpoint.

## 1. InvokeModel — Simplified (Single Checkpoint)

Add `cache_control` inside the content block:

```python
{
    "type": "text",
    "text": "<static content>",
    "cache_control": {"type": "ephemeral", "ttl": "5m"}
}
```

````python
def invoke_model_simplified():
    request_body = {
        "anthropic_version": "bedrock-2023-05-31",
        "max_tokens": 512,
        "messages": [{
            "role": "user",
            "content": [
                {
                    "type": "text",
                    "text": SPACE_CONTEXT,
                    "cache_control": {"type": "ephemeral", "ttl": CACHE_TTL}
                },
                {
                    "type": "text",
                    "text": QUESTION
                }
            ]
        }]
    }

    response = bedrock.invoke_model(modelId=MODEL_ID, body=json.dumps(request_body))
    return json.loads(response["body"].read()).get("usage", {})

# Request 1 — cache write
usage1 = invoke_model_simplified()
print("Request 1 (cache write expected):")
print(json.dumps(usage1, indent=2))

time.sleep(1)

# Request 2 — cache read
usage2 = invoke_model_simplified()
print("\nRequest 2 (cache read expected):")
print(json.dumps(usage2, indent=2))
````

## 2. InvokeModel — Multiple Checkpoints

Each section gets its own `cache_control` block.

````python
def invoke_model_multiple():
    request_body = {
        "anthropic_version": "bedrock-2023-05-31",
        "max_tokens": 512,
        "messages": [{
            "role": "user",
            "content": [
                {"type": "text", "text": SPACE_SECTION_1, "cache_control": {"type": "ephemeral", "ttl": CACHE_TTL}},
                {"type": "text", "text": SPACE_SECTION_2, "cache_control": {"type": "ephemeral", "ttl": CACHE_TTL}},
                {"type": "text", "text": SPACE_SECTION_3, "cache_control": {"type": "ephemeral", "ttl": CACHE_TTL}},
                {"type": "text", "text": SPACE_SECTION_4, "cache_control": {"type": "ephemeral", "ttl": CACHE_TTL}},
                {"type": "text", "text": QUESTION}
            ]
        }]
    }

    response = bedrock.invoke_model(modelId=MODEL_ID, body=json.dumps(request_body))
    return json.loads(response["body"].read()).get("usage", {})

# Request 1 — cache write
usage1 = invoke_model_multiple()
print("Request 1 (cache write expected):")
print(json.dumps(usage1, indent=2))

time.sleep(1)

# Request 2 — cache read
usage2 = invoke_model_multiple()
print("\nRequest 2 (cache read expected):")
print(json.dumps(usage2, indent=2))
````

## 3. InvokeModelWithResponseStream — Simplified

Same `cache_control` syntax with streaming. Cache metrics arrive across multiple chunk types:

```python
for event in response["body"]:
    chunk = json.loads(event["chunk"]["bytes"])
    if chunk["type"] == "message_start":
        # chunk["message"]["usage"] → cache_creation_input_tokens, cache_read_input_tokens
    elif chunk["type"] == "message_delta":
        # chunk["usage"] → output_tokens
```

````python
def invoke_model_stream_simplified():
    request_body = {
        "anthropic_version": "bedrock-2023-05-31",
        "max_tokens": 512,
        "messages": [{
            "role": "user",
            "content": [
                {
                    "type": "text",
                    "text": SPACE_CONTEXT,
                    "cache_control": {"type": "ephemeral", "ttl": CACHE_TTL}
                },
                {
                    "type": "text",
                    "text": QUESTION
                }
            ]
        }]
    }

    response = bedrock.invoke_model_with_response_stream(
        modelId=MODEL_ID,
        body=json.dumps(request_body)
    )

    text = ""
    usage_data = {}
    for event in response["body"]:
        chunk = json.loads(event["chunk"]["bytes"])
        if chunk["type"] == "content_block_delta":
            text += chunk["delta"].get("text", "")
        elif chunk["type"] == "message_delta":
            if "usage" in chunk:
                usage_data.update(chunk["usage"])
        elif chunk["type"] == "message_start":
            if "message" in chunk and "usage" in chunk["message"]:
                usage_data.update(chunk["message"]["usage"])

    return usage_data

# Request 1 — cache write
usage1 = invoke_model_stream_simplified()
print("Request 1 (cache write expected):")
print(json.dumps(usage1, indent=2))

time.sleep(1)

# Request 2 — cache read
usage2 = invoke_model_stream_simplified()
print("\nRequest 2 (cache read expected):")
print(json.dumps(usage2, indent=2))
````

## API Syntax Comparison

| Feature | InvokeModel API (Anthropic) | Converse API |
|---|---|---|
| Cache marker | `"cache_control": {"type": "ephemeral", "ttl": "5m"}` inside content block | `{"cachePoint": {"type": "default"}}` as standalone block |
| Usage response keys | `cache_creation_input_tokens`, `cache_read_input_tokens` | `cacheWriteInputTokens`, `cacheReadInputTokens` |
| `cacheDetails` (per-TTL breakdown) | No | Yes |
| Streaming usage location | `message_start` and `message_delta` chunks | `metadata` event |

## Conclusion

This notebook demonstrated message content caching using the Anthropic-specific InvokeModel syntax:

- `cache_control` is embedded inside each content block (unlike the standalone `cachePoint` in Converse)
- Streaming cache metrics are split across `message_start` and `message_delta` chunks
- The first request writes to cache (`cache_creation_input_tokens > 0`), subsequent requests read from cache (`cache_read_input_tokens > 0`)
